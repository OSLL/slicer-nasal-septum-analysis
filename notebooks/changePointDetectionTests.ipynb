# Тестирование разных алгоритмов

# 1. Linearly penalized segmentation (Pelt)

```python
def pelt():
    sigma = 80 # noise standard deviation
    
    model = "l2"  # l2, cosine
    algo = rpt.Pelt(model=model, min_size=MIN_SIZE).fit(maxIntensities)
    
    return algo.predict(pen=15000)

bkps = pelt()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

## 2. Kernel change point detection

### 2.1) Linear kernel (linear)

```python
def kernel_l2():
    sigma = 350  # noise standard deviation
    
    # Тут linear - это на самом деле l2
    algo = rpt.KernelCPD(kernel="linear", min_size=MIN_SIZE).fit(maxIntensities)
    
    # print(np.log(contourSize) * DIMENSIONS * sigma ** 2)
    return algo.predict(pen=15000)

bkps = kernel_l2()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

### 2.2) Gaussian kernel (rbf)

```python
def kernel_rbf():
    algo = rpt.KernelCPD(kernel="rbf", min_size=MIN_SIZE).fit(maxIntensities)
    return algo.predict(pen=1)

bkps = kernel_rbf()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

### 2.3) Cosine kernel

```python
def kernel_cosine():
    algo = rpt.KernelCPD(kernel="cosine", min_size=30).fit(maxIntensities)
    return algo.predict(pen=0.000000001)

bkps = kernel_cosine()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

## 3. Binary segmentation

```python
def binary():
    sigma = 80 # noise standard deviation
    
    model = "l2"  # l2, cosine
    algo = rpt.Binseg(model=model, min_size=MIN_SIZE).fit(maxIntensities)
    
    return algo.predict(epsilon=150000)
    # return algo.predict(epsilon=250)
    # return algo.predict(pen=100)

bkps = binary()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

## 4. Bottom-up segmentation

```python
def bottom_up():
    sigma = 6  # noise standard deviation
    
    model = "l2"  # l2, cosine
    algo = rpt.BottomUp(model=model, min_size=MIN_SIZE).fit(maxIntensities)
    
    return algo.predict(epsilon=1400000)
    # return algo.predict(pen=np.log(contourSize) * DIMENSIONS * sigma ** 2)

bkps = bottom_up()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

## 5. Window-based change point detection

```python
def window():
    sigma = 103 # noise standard deviation

    model = "l2" # l2, cosine
    algo = rpt.Window(width=MIN_SIZE*2, model=model, min_size=MIN_SIZE).fit(maxIntensities)
    
    return algo.predict(epsilon=1)
    # return algo.predict(pen=np.log(contourSize) * DIMENSIONS * sigma ** 2)

bkps = window()
# print(len(bkps))
# 
# rpt.show.display(maxIntensities, bkps, figsize=chartSize)
# plt.show()
```

# Тест

```python
import time

COUNT_LAUNCHES = 1000

def test_time(func):
    results = np.empty(COUNT_LAUNCHES)
    for i in range(COUNT_LAUNCHES):
        start_time = time.time()
        func()
        results[i] = time.time() - start_time
    
    # вычисляем среднее время
    avg_time = np.mean(results)
    print(f"Среднее время выполнения функции {func.__name__}: {avg_time * 1000.0:.2f} ms")

test_time(pelt)
test_time(kernel_l2)
test_time(binary)
test_time(bottom_up)
test_time(window)
```
